```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Funções para cálculos de Forex

def calculate_pips(initial_price, final_price):
    return (final_price - initial_price) * 10000

def calculate_pipets(initial_price, final_price):
    return (final_price - initial_price) * 100000

def calculate_lot_value(lot_type, unit_price):
    lot_sizes = {'standard': 100000, 'mini': 10000, 'micro': 1000}
    return lot_sizes[lot_type] * unit_price

def calculate_margin_required(position_size, leverage):
    return position_size / leverage

def calculate_spread(ask_price, bid_price):
    return (ask_price - bid_price) * 10000
```

```python
# Dados de exemplo
initial_price = 1.1050
final_price = 1.1051
ask_price = 1.1050
bid_price = 1.1048
unit_price = 1.1050
position_size = 100000  # standard lot
leverage = 100
```

```python
# Cálculos
pips = calculate_pips(initial_price, final_price)
pipets = calculate_pipets(initial_price, final_price)
lot_value_standard = calculate_lot_value('standard', unit_price)
lot_value_mini = calculate_lot_value('mini', unit_price)
lot_value_micro = calculate_lot_value('micro', unit_price)
margin_required = calculate_margin_required(position_size, leverage)
spread = calculate_spread(ask_price, bid_price)
```

```python
# Criação de tabelas com Pandas
data = {
    'Concept': ['Pips', 'Pipets', 'Lot Value (Standard)', 'Lot Value (Mini)', 'Lot Value (Micro)', 'Margin Required', 'Spread'],
    'Value': [pips, pipets, lot_value_standard, lot_value_mini, lot_value_micro, margin_required, spread]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Exibição da tabela
print(df)
```

```python
# Criação de gráficos com Matplotlib
fig, ax = plt.subplots()
# Gráfico de Barras
df.plot(kind='bar', x='Concept', y='Value', ax=ax, legend=False)
ax.set_ylabel('Value')
ax.set_title('Forex Calculations')
# Ajuste para exibir melhor os rótulos
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
# Exibição do gráfico
plt.show()
```
