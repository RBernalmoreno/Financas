```python
# Caminhos de destino locais
destination_paths = {
    "calls.txt": r'C:\Users\ROGÉRIO\AppData\Roaming\MetaQuotes\Terminal\Common\Files\Alien\calls.txt',
    "calls_compra.txt": r'C:\Users\ROGÉRIO\AppData\Roaming\MetaQuotes\Terminal\Common\Files\Alien\calls_compra.txt',
    "calls_venda.txt": r'C:\Users\ROGÉRIO\AppData\Roaming\MetaQuotes\Terminal\Common\Files\Alien\calls_venda.txt'
}
```

```python
def download_file(url, local_path):
    response = requests.get(url)
    if response.status_code == 200:
        with open(local_path, 'wb') as f:
            f.write(response.content)
        print(f'Arquivo {os.path.basename(local_path)} copiado em {time.ctime()}')
    else:
        print(f"Erro ao baixar {os.path.basename(local_path)}: Status code {response.status_code}")
```

```python
# Loop infinito para verificar e copiar os arquivos a cada 30 segundos
while True:
    try:
        for file_name, url in urls.items():
            local_path = destination_paths[file_name]
            download_file(url, local_path)
        
        # Aguardar 30 segundos antes de verificar novamente
        time.sleep(30)

    except Exception as e:
        print(f"Ocorreu um erro: {e}")
```
