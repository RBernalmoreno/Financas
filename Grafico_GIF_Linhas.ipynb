```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from PIL import Image
```

```python
# Configurar dados fictícios para investimentos (crescimento ascendente)
np.random.seed(0)
time = np.arange(0, 5, 0.1)
investment = np.cumsum(np.random.normal(0.5, 0.2, size=time.size))  # Crescimento ascendente
fig, ax = plt.subplots()
ax.set_xlim(0, 5)
ax.set_ylim(0, np.max(investment) * 1.1)
line, = ax.plot(time, investment, 'g-')  # 'g-' para linha verde
```

```python
# Função de atualização para o gráfico
def update(num, time, investment, line):
    line.set_data(time[:num], investment[:num])
    return line,

ani = animation.FuncAnimation(fig, update, frames=len(time), fargs=[time, investment, line], interval=100)
```

```python
# Salvar a animação como GIF
ani.save('C:/Temp/investment_growth3.gif', writer='pillow')
```
